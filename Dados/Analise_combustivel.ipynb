```python
import matplotlib.pyplot as plt

produto_escolhido = "GASOLINA C COMUM"
filtro = df['PRODUTO'] == produto_escolhido
df_filtrado = df[filtro]

plt.figure(figsize=(12, 6))
plt.plot(df_filtrado['DATA INICIAL'], df_filtrado['PREÇO MÉDIO DE DISTRIBUIÇÃO'], marker='o', linestyle='-', color='b')
plt.title(f'Preço Médio de Distribuição para {produto_escolhido}')
plt.xlabel('Data Inicial')
plt.ylabel('Preço Médio de Distribuição')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='PRODUTO', y='PREÇO MÉDIO DE DISTRIBUIÇÃO', data=df, palette='pastel')
plt.title('Distribuição de Preços Médios de Distribuição por Produto')
plt.xlabel('Produto')
plt.ylabel('Preço Médio de Distribuição')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Contar a frequência de cada categoria na coluna "TP_COR_RACA"
cor_raca_counts = df['PRODUTO'].value_counts()

# Definir cores para as categorias
cores = ['lightcoral', 'lightblue', 'lightgreen', 'lightgray', 'gold', 'lightcyan']

# Criar uma figura e eixos
fig, ax = plt.subplots()

# Criar o gráfico de pizza
ax.pie(cor_raca_counts, labels=cor_raca_counts.index, colors=cores, autopct='%1.1f%%')

# Adicionar título
ax.set_title('Distribuição de Produtos')

# Mostrar o gráfico plotado
plt.show()

```
