```python
import matplotlib.pyplot as plt
import pandas as pd


# Contar a frequência de cada categoria na coluna "TP_SEXO"
proporcao_feminino = df['TP_SEXO'].value_counts()


# Definir cores para as categorias
cores = ['lightblue', 'lightcoral']

# Criar uma figura e eixos
fig, ax = plt.subplots()

# Criar o gráfico de pizza
ax.pie(proporcao_feminino,  labels=proporcao_feminino.index, colors=cores, autopct='%1.1f%%')

# Adicionar título
ax.set_title('Distribuição de Produtos')

# Mostrar o gráfico plotado
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd


# Contar a frequência de cada categoria na coluna "TP_COR_RACA"
cor_raca_counts = df['TP_COR_RACA'].value_counts()

# Definir cores para as categorias
cores = ['lightcoral', 'lightblue', 'lightgreen', 'lightgray', 'gold', 'lightcyan']

# Criar uma figura e eixos
fig, ax = plt.subplots()

# Criar o gráfico de pizza
ax.pie(cor_raca_counts, labels=cor_raca_counts.index, colors=cores, autopct='%1.1f%%')

# Adicionar título
ax.set_title('Distribuição de Raça/Etnia')

# Mostrar o gráfico plotado
plt.show()

```
